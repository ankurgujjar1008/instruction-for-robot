```python
import turtle
import speech_recognition as sr
import re
import os
```

```python
# Create a speech recognition object
r = sr.Recognizer()
```

```python
def extract_intent(text):
    # Define regular expressions for the intents
    forward_pattern = r"move forward (\d+) meters?"
    backward_pattern = r"move backward (\d+) meters?"
    left_pattern = r"turn left (\d+)"
    right_pattern = r"turn right (\d+)"
    speed_up_pattern = r"speed up (\d+)%"
    speed_down_pattern = r"speed down by (\d+)%"
    stop_pattern = r"stop"

    # Search for the patterns in the text
    forward_match = re.search(forward_pattern, text, re.IGNORECASE)
    backward_match = re.search(backward_pattern, text, re.IGNORECASE)
    left_match = re.search(left_pattern, text, re.IGNORECASE)
    right_match = re.search(right_pattern, text, re.IGNORECASE)
    speed_up_match = re.search(speed_up_pattern, text, re.IGNORECASE)
    speed_down_match = re.search(speed_down_pattern, text, re.IGNORECASE)
    stop_match = re.search(stop_pattern, text, re.IGNORECASE)

    # Extract the values from the matches
    if forward_match:
        forward_distance = int(forward_match.group(1))
    else:
        forward_distance = None

    if backward_match:
        backward_distance = int(backward_match.group(1))
    else:
        backward_distance = None

    if left_match:
        left_angle = int(left_match.group(1))
    else:
        left_angle = None

    if right_match:
        right_angle = int(right_match.group(1))
    else:
        right_angle = None

    if speed_up_match:
        speed_up_percent = int(speed_up_match.group(1))
    else:
        speed_up_percent = None
    
    if speed_down_match:
        speed_down_percent = int(speed_down_match.group(1))
    else:
        speed_down_percent = None

    if stop_match:
        stop = True
    else:
        stop = False

    # Return the extracted intent
    return {
        "forward": forward_distance,
        "backward": backward_distance,
        "left": left_angle,
        "right": right_angle,
        "speed_up": speed_up_percent,
        "speed_down": speed_down_percent,
        "stop": stop
    }

```

```python
# Specify the audio file
audio_dir = 'C:/Users/hp/Desktop'
```

```python

import turtle

# Create a turtle screen and set its background color
screen = turtle.Screen()
screen.setup(width=800, height=600)

# Create a turtle object
robot = turtle.Turtle()
robot.shape("turtle")
robot.color("blue")

# Set the turtle's movement speed
robot.speed('slowest')

def move_forward(distance):
    robot.forward(distance)

def turn_left(angle):
    robot.left(angle)

def turn_right(angle):
    robot.right(angle)

def move_backward(distance):
    robot.backward(distance)

def stop():
    print("Stopping")

for filename in os.listdir(audio_dir):
    if filename.endswith(".wav"):
        audio_file = os.path.join(audio_dir, filename)

        # Use the audio file as the audio source
        with sr.AudioFile(audio_file) as source:
            audio = r.record(source)

            # Recognize speech in the audio
            try:
                text = r.recognize_google(audio, language='en-US')
                print("Recognized text from file", filename, ":", text)

                # Extract the intent from the recognized text
                intent = extract_intent(text)
                print("Extracted intent:", intent)

                # Move the turtle based on the intent
                if intent["forward"] is not None:
                    move_forward(intent["forward"])
                if intent["backward"] is not None:
                    move_backward(intent["backward"])
                if intent["left"] is not None:
                    turn_left(intent["left"])
                if intent["right"] is not None:
                    turn_right(intent["right"])

            except sr.UnknownValueError:
                print("Speech recognition could not understand audio from file", filename)
            except sr.RequestError as e:
                print("Could not request results from Google Speech Recognition service for file", filename, "; {0}".format(e))

# Execute the commands
move_forward(30)
turn_left(32)
move_forward(55)
turn_right(45)
move_forward(30)
turn_right(20)
move_backward(20)
turn_left(90)
move_forward(40)
stop()

# Keep the window open
turtle.done()

```
